```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2
```

### Basic image manipulation

```python
#Convert to RGB
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
#Display image in grayscale
img_gray = cv2.imread('../../opencv_udemy/DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
img_gray.shape
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
#Resize
new_img = cv2.resize(img2, (250, 500))
plt.imshow(new_img)
```

```python
#Resize with ratio
img3 = cv2.resize(img2, (0, 0), img2, 0.5, 0.5)
img3.shape
```

```python
plt.imshow(img3)
```

```python
#Flip image
img4 = cv2.flip(img2, -1)
plt.imshow(img4)
```

### Drawing figures

```python
img = np.zeros((512, 512, 3), dtype=np.int16)
```

```python
plt.imshow(img)
```

```python
cv2.rectangle(img, (15, 15), (200, 300), (0, 255, 0), 10)
plt.imshow(img)
```

```python
#daw a circle
cv2.circle(img=img, center=(350, 400), radius=50, color=(0, 0, 255), thickness=10)
plt.imshow(img)
```

```python
#daw a line
cv2.line(img=img, pt1=(150, 100), pt2=(350, 400), color=(255, 0, 0), thickness=10)
plt.imshow(img)
```

```python
#Write text
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, text='Test', org=(250, 200), fontFace=font, fontScale=4, color=(255,255,255), thickness=2)
plt.imshow(img)
```

```python
#Draw a poligon
img2 = np.zeros((512, 512, 3), dtype=np.int16)
plt.imshow(img2)
```

```python
vertices = np.array([[50, 100], [75, 450], [450,450], [300, 200]], dtype=np.int32)
#change dimensions
vertices = vertices.reshape((-1, 1, 2))
vertices.shape
```

```python
cv2.polylines(img2, [vertices], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(img2)
```
